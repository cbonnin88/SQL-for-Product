```python
import polars as pl
import numpy as np
from datetime import datetime, timedelta
```

```python
num_users = 10000
num_stations = 500
num_trips = 50000
countries = ["France", "Germany", "Netherlands", "Spain", "Austria"]

print("Initializing data generation for EV Charging App...")

# ---------------------------------------------------------
# 1. Users Dataset
# ---------------------------------------------------------
# Gender setup
genders = ["Male", "Female", "Non-Binary"]
gender_probs = [0.55, 0.43, 0.02]

# Age setup: Beta distribution (a=2, b=5) creates a right-skewed curve (younger).
# We scale the output (which is between 0 and 1) to our desired age range of 18 to 75.
raw_ages = np.random.beta(a=2, b=5, size=num_users)
scaled_ages = 18 + (raw_ages * (75 - 18))
ages = np.round(scaled_ages).astype(int)

# Subscription setup
tiers = ["Free", "Basic", "Premium"]
tier_probs = [0.60, 0.25, 0.15]
price_map = {"Free": 0.00, "Basic": 2.99, "Premium": 5.99}

# Generate tiers based on probabilities, then map them to prices
assigned_tiers = np.random.choice(tiers, num_users, p=tier_probs)
assigned_prices = [price_map[tier] for tier in assigned_tiers]

users_df = pl.DataFrame({
    "user_id": np.arange(1, num_users + 1),
    "signup_date": [datetime(2023, 1, 1) + timedelta(days=np.random.randint(0, 365)) for _ in range(num_users)],
    "country": np.random.choice(countries, num_users),
    "gender": np.random.choice(genders, num_users, p=gender_probs),
    "age": ages,
    "subscription_tier": assigned_tiers,
    "subscription_price": assigned_prices
})

print("Writing users.csv...")
users_df.write_csv("users.csv")

# ---------------------------------------------------------
# 2. Stations Dataset
# ---------------------------------------------------------
stations_df = pl.DataFrame({
    "station_id": np.arange(1, num_stations + 1),
    "country": np.random.choice(countries, num_stations),
    "latitude": np.random.uniform(36.0, 53.0, num_stations),
    "longitude": np.random.uniform(-9.0, 14.0, num_stations),
    "plug_count": np.random.randint(2, 10, num_stations)
})

print("Writing stations.csv...")
stations_df.write_csv("stations.csv")

# ---------------------------------------------------------
# 3. Trips Dataset
# ---------------------------------------------------------
trips_df = pl.DataFrame({
    "trip_id": np.arange(1, num_trips + 1),
    "user_id": np.random.randint(1, num_users + 1, num_trips),
    "start_time": [datetime(2024, 1, 1) + timedelta(minutes=np.random.randint(0, 500000)) for _ in range(num_trips)],
    "distance_km": np.random.uniform(10.0, 500.0, num_trips)
})

print("Writing trips.csv...")
trips_df.write_csv("trips.csv")

# ---------------------------------------------------------
# 4. Charge Events Dataset
# ---------------------------------------------------------
charge_events_df = pl.DataFrame({
    "event_id": np.arange(1, num_trips + 1),
    "trip_id": trips_df["trip_id"],
    "station_id": np.random.randint(1, num_stations + 1, num_trips),
    "kwh_charged": np.random.uniform(5.0, 60.0, num_trips),
    "cost_eur": np.random.uniform(2.0, 45.0, num_trips),
    "is_successful": np.random.choice([True, False], num_trips, p=[0.95, 0.05])
})

print("Writing charge_events.csv...")
charge_events_df.write_csv("charge_events.csv")

print("Phase 0 Complete: All 4 CSV files generated successfully!")
```
